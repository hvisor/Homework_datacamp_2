5. นำข้อ 3 และ 4 มาเปรียบเวลากัน โดยใช้ฟังก์ชั่น time.time( ) เพื่อจับเวลา แสดงผลลัพธ์ในรูปแบบมิลลิวินาที



```
import random
import time
iMin, iMax = 10, 99

randomList = []
# Set a length of the list to 20
for i in range(1, 20):
    # any random numbers from 10 to 99
    randomList.append((iMax - iMin)*random.random()+iMin)
print(randomList)
# print(type(randomList))
```

```
npArray = np.array(randomList)
# print(npArrayRandom)
# print(type(npArray))

time1 = time.time()
x = npArray**2
print(x)
time2 = time.time()
timeOfNP = time2-time1

time1 = time.time()
y = [x ** 2 for x in randomList]
# print(type(y))
print(y)
time2 = time.time()
timeOfList = time2-time1
print('NP Used time ',timeOfNP)
print('List Used time ',timeOfList)
```

20.เขียนโปรแกรมรับค่าจากผู้ใช้ให้ป้อนเลข 9 ตัว แต่ละตัวต้องไม่ซ้ำกันและมีค่าอยู่ในช่วง 10-100 นำเลข 9 ตัวนั้นมาทำเป็น matrix 
      3x3 และ transport และหาเลขอยู่ตำแหน่งที่ (2,2)

```

lst = []
# print (lst)
vInput2 = 0
while (len(lst)<=9):
    vInput2 = int(input('Input ที่ '+str(len(lst)+1)+' (10 - 100) :'))
    if( vInput2 >= 10 and vInput2 <= 100):
        if (vInput2 not in lst):
            lst.append(vInput2)
            if(len(lst)==9):
                break;

print(lst)
```

```
arr2D = np.array(lst)
print(arr2D)
arr2D = arr2D.reshape(3,3)
print(arr2D)
arr2D = arr2D.T
print(arr2D)
print(arr2D[2,2])
```
